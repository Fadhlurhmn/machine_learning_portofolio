```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ====== 1. Load Data ======
df = pd.read_csv("/kaggle/input/tweet-sentiment-classification-dataset/tweet_sentiment.csv")  # Ganti dengan nama file kamu
print(f"üìä Dataset loaded: {df.shape[0]} rows √ó {df.shape[1]} columns")

# ====== 2. Basic Info ======
print("\nüìã Basic Info:")
print(df.info())

# ====== 3. Missing Values & Duplicate Rows Check ======
print("\nüîç Missing Values:")
missing = df.isnull().sum()
print(missing[missing > 0])

print("\nüîç Duplicate Rows:")
duplicate = df.duplicated().sum()
print(duplicate)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Set Target
X = df.drop(columns=['sentiment'])
y = df['sentiment']

# Deteksi kolom kategorikal dan numerik dari X
cat_cols = X.select_dtypes(include='object').columns.tolist()

# Pipeline untuk Kategorikal + Handling Missing Values
categorical_pipeline = Pipeline([
    ('encoder', OneHotEncoder(handle_unknown='ignore')) # bisa diganti untuk LabelEncoder
		# bisa ditambah singleimputer untuk mengisi nilai kosong
])

# Gabung dengan ColumnTransformer
preprocessor = ColumnTransformer([
    ('cat', categorical_pipeline, cat_cols)
])

# Full Pipeline Model
pipeline = Pipeline([
    ('preprocessing', preprocessor),
    ('regressor', RandomForestClassifier()) # model bisa diganti regresi atau yg lain dan model yang digunakan
])

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Fit pipeline
pipeline.fit(X_train, y_train)

# Prediksi
y_pred = pipeline.predict(X_test)

# Evaluasi model
report = classification_report(y_test, y_pred)
print(report)
```
